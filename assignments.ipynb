```python
#1 to 4
def check_num(n):
    if n > 0:
        if n % 2 == 0:
            return f"The {n} number is positive and even"
        return f"The {n} number is positive and odd"
    elif n < 0:
        if n % 2 == 0:
            return f"The {n} number is negative and even"
        return f"The {n} number is negative and odd"
    return f"The entered object is 0, neither positive nor negative."


user_input = int(input("Enter your number: "))

print(check_num(user_input))
```
