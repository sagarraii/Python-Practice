```python
# 6
# nested Exception Handling, create function take num as input convert to int
def nested_exception_handling(n):
    try:
        try:
            num = int(n)
        except ValueError as e:
            print(f"Error: {e}")
            num = None
        finally:
            print("Execution Completed.")
        if num is not None:
            try:
                result = 10/num
            except ZeroDivisionError as e:
                print(f"Division Error: {e}")
                result = None
            finally:
                print("Division attempt completed.")
            return result
    finally:
        print("Overall Execution Completed.")

# Test
print(nested_exception_handling('0'))  # None
print(nested_exception_handling('a'))  # None
print(nested_exception_handling('2'))  # 5.0
```

```python
#7 Exception Handling in list operatioms

def get_list_elements(lst, index):
    try:
        elements = lst[index]
    except IndexError as e:
        print(f"Error: {e}")
        elements = None
    finally:
        print("Executio Completed.")
    return elements

# Test
lst = [1, 2, 3, 4, 5]
print(get_list_elements(lst, 2))  # 3
print(get_list_elements(lst, 10))  # None
```

```python
# 8 Exception handling in Network Operations

import requests

def read_url(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.text
    except requests.RequestException as e:
        print(f"Network Error: {e}")
        return None
    finally:
        print("Execution Completed.")


# Test
print(read_url('https://jsonplaceholder.typicode.com/posts/1'))
print(read_url('https://nonexistent.url'))        
```

```python
#9 Exception Handling in JSON Parsing

import json

def parse_json(json_string):
    try:
        data = json.loads(json_string)
        return data
    except json.JSONDecodeError as e:
        print(f"JSON Error: {e}")
        return None
    finally:
        print("Execution Complete.")

# Test
print(parse_json('{"name": "John", "age": 30}'))  # {'name': 'John', 'age': 30}
print(parse_json('Invalid JSON'))  # None
```

```python
# 10 Custom Exception Handling

class NegativeNumberError (Exception):
    pass

def negative_num(lst):
    try:
        for num in lst:
            if num < 0:
                raise NegativeNumberError(f"Negative number found: {num}")
    except NegativeNumberError as e:
        print(f"Error: {e}")
    finally:
        print("Execution Completed.")

negative_num([1, -2, 3, 4])
negative_num([1, 2, 3, 4]) 
```

```python
# 12 Exception Handling in class method

class Calculator:
    def divide(self, a , b):
        try:
            result = a/b
        except ZeroDivisionError as e:
            print(f"Error: {e}")
            result = None
        finally:
            print("Execution Completed.")
        return result
    
calc = Calculator()
print(calc.divide(10, 2))  # 5.0
print(calc.divide(10, 0))  # None
```

```python
# 13 Exception Handling in Data Conversion

def convert_to_integers(lst):
    integer = []
    try:
        for item in lst:
            integer.append(int(item))
    except ValueError as e:
        print(f"Error: {e}")
    finally:
        print("Execution Completed.")
    return integer

# Test
print(convert_to_integers(['1', '2', 'three', '4']))  # None
print(convert_to_integers(['1', '2', '3', '4']))  # [1, 2, 3, 4]
```

```python
def list_comprehensions(lst):
    try:
        integer = [int(items) for items in lst]
    except ValueError as e:
        print(f"Error: {e}")
        integer = None
    finally:
        print("Eecution Completed")
    return integer

# Test
print(list_comprehensions(['1', '2', 'three', '4']))  # None
print(list_comprehensions(['1', '2', '3', '4']))  # [1, 2, 3, 4]
```

```python
# 15 Exception Handling with file writing

def lst_of_strings(filename, strings):
    try:
        file = open(filename, 'w')
        for string in strings:
            file.write(string + '\n')
    except IOError as e:
        print(f"Error: {e}")
    finally:
        try:
            file.close()
        except NameError:
            pass

# Test
lst_of_strings(['Hello', 'World'], 'output.txt')
```

```python
def write_strings_to_file(strings, filename):
    try:
        file = open(filename, 'w')
        for string in strings:
            file.write(string + '\n')
    except IOError as e:
        print(f"Error: {e}")
    finally:
        try:
            file.close()
        except NameError:
            pass

# Test
write_strings_to_file(['Hello', 'World'], 'output.txt')
```
