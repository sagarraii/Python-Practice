```python
def fibo(n, memo={}):
    if n in memo:
        return memo[n]
    if n <=1:
        return n
    memo[n]= fibo(n-1, memo) + fibo(n-2, memo)
    return memo[n]

print(fibo(6))
```

```python
def default_arg(a, b=None):
    if b is None:
        b = {}

    #d[key] = Value
    b[a] = a**2
    return b

print(default_arg(4))
```

```python
def filter_integers(**kwargs):
    return {k: v for k, v in kwargs.items() if isinstance(v, int)}

print(filter_integers(a=1, b='two', c=3, d=4.5))  # {'a': 1, 'c': 3}
print(filter_integers(x=10, y='yes', z=20))  # {'x': 10, 'z': 20}
```

```python
def apply_callback(callback, lst):
    return [callback(x) for x in lst]

print(apply_callback(lambda x: x**2, [1, 2, 3, 4, 5, 6, 7]))
```

```python
import time

def timer_decorator(function):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = function(*args, **kwargs)
        close_time = time.time()
        print(f"function {function.__name__} took {close_time - start_time} seconds to execute")
        print(result)
    return wrapper

@timer_decorator
def sqr(n):
    return sum(x**2 for x in range(n))

#test 
print(sqr(100000))
        
```

```python
def high_order_function(filter_fun, map_funct, lst):
    return [map_funct(x) for x in lst if filter_fun(x)]

#test
print(high_order_function(lambda x: x % 2 == 0, lambda x: x ** 2, [1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

```python
def compose(f, g):
    return lambda x: f(g(x))

# Test
f = lambda x: x + 1
g = lambda x: x * 2
h = compose(f, g)
print(h(3))  # 7
print(h(5))  # 11
```
