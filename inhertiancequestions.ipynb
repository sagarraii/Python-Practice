```python
# 1
# base class
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species


# derived class
class Dog(Animal):
    def __init__(self, name, species, breed):
        super().__init__(name, species)
        self.breed = breed

dog = Dog("bunny", "Canine", "Golden Retriever")
print(dog.name, dog.species, dog.breed)
```

```python
class Dog(Animal):
    def __init__(self, name, species, breed):
        super().__init__(name, species)
        self.breed = breed

    def __str__(self):
        return f"Dog(Name: {self.name}, Species: {self.species}, Breed: {self.breed})"

dog = Dog("bunny", "Canine", "Golden Retriever")
print(dog)
```

```python
class Dog(Animal):
    def __init__(self, name, species, breed):
        super().__init__(name, species)
        self.breed = breed

    def bark(self):
        print("Woof")

#object of the class
dog = Dog("bunny", "Canine", "Golden Retriever")
dog.bark()
```

```python
# 4
# Multiple inheritance 
# base class

class Walker:
    def walk(self):
        print("Man is Wlking")


# another base class
class Runner:
    def run(self):
        print("Man is running.")

# Derived class
class Athlete(Walker, Runner):
    pass

athlete = Athlete()
athlete.walk()
athlete.run()
```

```python
# 5
# Method Order Resolution (MOR) in multiple inheritance

class Athlete(Walker, Runner):
    def walk(self):
        print("Athlete is walking")
        super().walk()

athlete = Athlete()
athlete.walk()
```

```python
# 6 
# Multiple inheritance with Additional attribute

class Athlete(Walker, Runner):
    def __init__(self, training_hour):
        self.training_hour = training_hour

    def train(self):
        print(f"Athlete train's {self.training_hour} hour per day")

# Test 
athlete = Athlete(6)
athlete.train()
```

```python
# 7 
# Dimaond problem in multiple inheritance

# Base class
class A:
    def show(self):
        print("this is A")

# Derive class of A
class B(A):
    def show(self):
        print("B shows method")

# Derive class of A
class C(A):
    def show(self):
        print("c shows method")

# Derive class of B and C
class D(B, C):
    pass

# test

d = D()
d.show()
```

```python
# 8

class Shape():
    def __init__(self, color):
        self.color = color

class Circle(Shape):
    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius

circle = Circle('Red', 5)
print(circle.color, circle.radius)
```

```python
# 9
# Using `super()` in Multiple Inheritance

class Person:
    def __init__(self, name):
        self.name = name

class Employee:
    def __init__(self, employee_id):
        self.employee_id = employee_id

class Manager(Person, Employee):
    def __init__(self, name, employee_id):
        super().__init__(name)
        Employee.__init__(self, employee_id)

# Test
manager = Manager('John', 'M123')
print(manager.name, manager.employee_id)
```

```python
# 10
# Method overriding super()

class Vehicle:
    def start(self):
        print(" Vehicle starting...")

class Car(Vehicle):
    def start(self):
        print("Car is starting...")
        super().start()

car = Car()
car.start()
```

```python
# 11 
# mutiple inheritance with different method

class Flyer:
    def fly(self):
        print("Flying Message...")

class Swimmer:
    def swim(self):
        print("Swimming Message...")

class Superhero(Flyer, Swimmer):
    pass

superhero = Superhero()
superhero.fly()
superhero.swim()
```

```python
# 12
# Complex Multiple Inheritance

class Base1:
    def __init__(self, a):
        self.a = a

class Base2:
    def __init__(self, b):
        self.b = b

class Derived(Base1, Base2):
    def __init__(self, a, b, c):
        super().__init__(a)
        Base2.__init__(self, b)
        self.c = c

# Test
derived = Derived(1, 2, 3)
print(derived.a, derived.b, derived.c)
```

```python
# 13 Checking INstance type with inheritance

class Animal:
    pass

class Cat(Animal):
    pass

# Test
animal = Animal()
cat = Cat()
print(isinstance(animal, Animal))  # True
print(isinstance(cat, Animal))  # True
print(isinstance(cat, Cat))  # True
print(isinstance(animal, Cat))  # False
```

```python
# Polymorphism with Inheritance

class Bird:
    def speak(self):
        print("Bird Speak")

class Parrot(Bird):
    def speak(self):
        print("Parrot Speak")

class Penguin(Bird):
    def speak(self):
        print("penguin sound")


birds = [Parrot(), Penguin()]
for bird in birds:
    bird.speak()
```

```python
# 15

# Base Class
class Device:
    def __init__(self, brand):
        self.brand = brand

# Derived class of device
class Phone(Device):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

# Another Base Class
class Camera:
    def __init__(self, resolution):
        self.resolution = resolution

# Derived class of Phone and Camera
class Smartphone(Phone, Camera):
    def __init__(self, brand, model, resolution):
        Phone.__init__(self, brand, model)
        Camera.__init__(self, resolution)

# object of SmartPhone

smartphone = Smartphone('Apple', '16 PRO MAX', '50 MP')
print(smartphone.brand, smartphone.model, smartphone.resolution)
```
