**Question 4:** Write a Python program to find the largest of three numbers.

```python
# Your code here
def largestnum(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
    
print(largestnum(int(input("Enter first number: ")), int(input("Enter second number: ")), int(input("Enter third number: "))))
```

**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Your code here
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1) 

print(factorial(int(input("Enter a number: "))))
```

**Question 16:** Write a Python program to find the sum of the first n natural numbers.

```python
# Your code here
def sum_of_n(n):
    if n <= 0:
        return 0
    return n + sum_of_n(n - 1)

number = int(input("Enter a number: "))
print("The number entered is:", number)
print(f"The sum of first {number} natural numbers is:", sum_of_n(number))
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python
# Your code here
def leap_year(n):
    if n % 400 == 0:
        return "leap year"
    elif n % 100 == 0:
        return "not leap year"
    elif n % 4 == 0:
        return "leap year"
    else:
        return "not leap year"


year = int(input("Enter you year: "))
print(leap_year(year))
```
